```python
import numpy as np
import cv2 as cv
import skimage.color
import skimage.io
import matplotlib.pyplot as plt
```

```python
img = cv.imread("git.png")
img = cv.resize(img,(800,600))
s = img.shape

imgGray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
```

```python
hist = cv.calcHist([imgGray], [0], None, [256], [0,256])
plt.plot(hist)
plt.show()
```

```python
def calcComulative(data):
    comulative_hist = []
    temp = 0
    for i in range(0,255):
        temp = temp + data[i,0]
        comulative_hist.append(temp)
        
    return comulative_hist

plt.plot(calcComulative(hist))
plt.show()
```

```python
original = cv.imread('git.png')
original = cv.resize(img,(800,600))
s = original.shape

imgGray2 = cv.cvtColor(original, cv.COLOR_BGR2GRAY)
equ = cv.equalizeHist(imgGray2)
res = np.hstack((imgGray2,equ))
cv.imwrite('res.png',res)
```

```python
hist = cv.calcHist([equ], [0], None, [256], [0,256])
plt.plot(hist)
plt.show()
```
